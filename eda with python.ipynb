# checking for null values and imputing with mean and median


```python
purchase = purchase.replace({'M':1,'U':0})
purchase.head(10)
```

```python
########finding 
def var_summary(x):
    return pd.Series([x.count(),x.isnull().sum(),x.sum(),x.mean(),x.median(),x.std(),x.var(),x.std()/x.mean(),
                      x.min(),x.dropna().quantile(0.01),x.dropna().quantile(0.05),x.dropna().quantile(0.10),
                      x.dropna().quantile(0.25),x.dropna().quantile(0.50),x.dropna().quantile(0.75),x.dropna().quantile(0.90),
                     x.dropna().quantile(0.95),x.dropna().quantile(0.99),x.max()],
                    index = ['N','NMISS','SUM','MEAN','MEDIAN','STD','VAR','CV','MIN','P1','P5','P10','P25','P50','P75',
                             'P90','P95','P99','MAX'])
purchase.apply(var_summary)
```

```python
purchase['PRICE'] =purchase['PRICE'].fillna(purchase['PRICE'].mean()) 
purchase.apply(var_summary)
```

```python
purchase['PUR_VALUE']=purchase['PUR_VALUE'].fillna(purchase['PUR_VALUE'].median())
purchase['USER_INCOME']=purchase['USER_INCOME'].fillna(purchase['USER_INCOME'].median())
purchase.apply(var_summary)
```

```python
######removimg outliers from the variable user_income
upper = purchase['USER_INCOME'].quantile(0.75)+1.5*(purchase['USER_INCOME'].quantile(0.75)-purchase['USER_INCOME'].quantile(0.25))
```

```python
######removimg outliers from the variable user_income
lower = purchase['USER_INCOME'].quantile(0.75)-1.5*(purchase['USER_INCOME'].quantile(0.75)-purchase['USER_INCOME'].quantile(0.25))
```

```python
#####handiling outliers-method1
purchase['USER_INCOME'] = purchase['USER_INCOME'].clip_upper(upper)
purchase['USER_INCOME'] = purchase['USER_INCOME'].clip_lower(lower)
```

```python
var_names = [key for key in dict(purchase.dtypes)]
bp = PdfPages('BoxPlots Purchase Data.pdf')
for num_variable in var_names:
    upper = purchase[num_variable].quantile(0.75)+1.5*(purchase[num_variable].quantile(0.75)-purchase[num_variable].quantile(0.25))
    lower = purchase[num_variable].quantile(0.75)-1.5*(purchase[num_variable].quantile(0.75)-purchase[num_variable].quantile(0.25))
    purchase[num_variable] = purchase[num_variable].clip_upper(upper)
    purchase[num_variable] = purchase[num_variable].clip_lower(lower)
    fig,axes = plt.subplots(figsize = (10,4))
    sns.boxplot(y = num_variable,data = purchase)
    plt.title(str('boxplot of ')+ str(num_variable))
    bp.savefig(fig)
bp.close()
```
